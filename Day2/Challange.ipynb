# Grade of Total Marks

```python
Subject_Name = ['Maths','Python','SQL','DSA','ML']
Subject_Mark = ['','','','','']

for i in range(len(Subject_Mark)):
    print("Please Enter Marks for ",Subject_Name[i]," : ")
    x = float(input())

    Subject_Mark[i] = x

Total = 0

for i in range(len(Subject_Mark)):
    print(Subject_Name[i]," Marks are : ",Subject_Mark[i])
    Total = Total + Subject_Mark[i]

    if Subject_Mark[i] <= 18 :
        print("Sorry you did not clear the exam for ", Subject_Name[i])

print("Total Marks are : ", Total)

T = "P"

for i in range(len(Subject_Name[i])):
    if Subject_Mark[i] <= 18:
        T = "F"
        print("Sorry you have not cleared this exam for ", Subject_Name[i])

if T == "F":
    print("fail")
elif Total == 500:
    print("Grade : AA")
elif Total >= 400:
    print("Grade : BB")
elif Total >= 300:
    print("Grade : CC")
else:
    print("DD")
```

# Subject vise grades

```python
Subject_Name = ['Maths','Python','SQL','DSA','ML']
Subject_Mark = ['','','','','']
Grades = ['','','','','']

for i in range(len(Subject_Mark)):
    print("Please Enter Marks for ",Subject_Name[i]," : ")
    x = float(input())

    Subject_Mark[i] = x

Total = 0

for i in range(len(Subject_Mark)):
    if Subject_Mark[i] == 100:
        Grade = "AA"
    elif Subject_Mark[i] >= 80:
        Grade = "BB"
    elif Subject_Mark[i] >= 18:
        Grade = "CC"
    else:
        Grade = ""

    Grades[i] = Grade
    
    print(Subject_Name[i]," Marks are : ",Subject_Mark[i]," And your grades are : ",Grade)

    Total = Total + Subject_Mark[i]

print("Total Marks are : ", Total)

for i in range(len(Grades)) :
    if Grades[i] == "":
        print("Sorry! You did not Clear the exam for", Subject_Name[i])
    
    elif Total == 500:
        print("AA")
```

# Trying to simplify the code

```python
Num = ['','','','','']
sort = ['','','','','']

for i in range(len(Num)):
    print("Please enter number ",i," : ",)
    x = float(input())
    Num[i] = x

print(Num)

Max = 0

for j in range(0,5):
    for i in range(0,5):
        if Num[i] >= Max:
            Max = Num[i]

        if Max == Num[j] and Num[j] != 0:
                Num[j] = 0


print("Maximum Number is: ", Max)
print(Num)
print(sort)
```

# SImplification v2

```python
Num = ['','','','','']
sort = ['','','','','']

for i in range(len(Num)):
    print("Please enter number ",i," : ",)
    x = float(input())
    Num[i] = x

print(Num)

Max = 0

for k in range(0,5):
    for j in range(0,5):
        for i in range(0,5):
            if Num[i] >= Max:
                Max = Num[i]

            if Max == Num[j]:
                Num[j] = 0
        
    sort[k] = Max
    Max = 0


print("Maximum Number is: ", Max)
print(Num)
print(sort)
```

# Calendar printer using nnumber

# Calculator using Match

```python
print("Please enter number 1 : ")
x = float(input())

print("please enter number 2 : ")
y = float(input())

print("Choose the arithmatic operation : ")
print("Enter 1 for Addition : \nEnter 2 for Substraction : \nEnter 3 for Multiplication : \nEnter 4 for Devision : ")
a = int(input())

match a:
    case 1:
        print("Addition of the two numbers is : ", x + y)
    case 2:
        print("Subtraction of the two numbers is : ", x - y)
    case 3:
        print("Multiplication of the two number is : ", x * y)
    case 4:
        print("Devision of the two numbers is : ", x / y)
    case _:
        print("Not an valid option")
```

# Grading according to marks

```python
Subject_Name = ['Maths','Python','SQL','DSA','ML']
Subject_Mark = ['','','','','']

for i in range(len(Subject_Mark)):
    print("Please Enter Marks for ",Subject_Name[i]," : ")
    x = float(input())

    Subject_Mark[i] = x

Total = 0

for i in range(len(Subject_Mark)):
    print(Subject_Name[i]," Marks are : ",Subject_Mark[i])
    Total = Total + Subject_Mark[i]

print("Total Marks are : ", Total)

Avg = (Subject_Mark[0] + Subject_Mark[1] + Subject_Mark[2] + Subject_Mark[3] + Subject_Mark[4])/5
print("Your Average Marks are : ", Avg)

if Avg < 50:
    print("Grade : F")
elif Avg >= 50 and Avg < 60:
    print("Grade : D")
elif Avg >= 60 and Avg < 70:
    print("Grade : C")
elif Avg >= 70 and Avg < 80:
    print("Grade : B")
elif Avg >= 80 and Avg < 90:
    print("Grade : A")
else:
    print("Grade : A+")
```

```python
def doSum(a,b,c):
    d = a + b + c
    return d

def diAvg(Total):
    a = Total/3
    return a

no1 = int(input("Enter no : 1"))
no2 = int(input("Enter no : 2"))
no3 = int(input("Enter no : 3"))

sumans = doSum(no1,no2,no3)
print(sumans)

avgans = diAvg(sumans)
print(avgans)
```

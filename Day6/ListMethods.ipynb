# Sort

```python
Fru = ["apple","chery","deova","banana","guava"]

nu = [1,10,3,2,0]

print(Fru)
Fru.sort()
print(Fru)

print(nu)
nu.sort()
print(nu)
```

# Reverse

```python
Fru.reverse()

print(Fru)
```

# Sorted

```python
print(sorted(nu))
```

# Max & Min

```python
print(max(nu))
print(min(nu))
```

# Sum

```python
print(sum(nu))
```

# Count

```python
cnt = ['python','html','css','js','sql','python','python','html']
print(cnt)

LiCnt = cnt.count('python')
LiCnt2 = cnt.count('js')
LiCnt3 = cnt.count('html')

print(LiCnt)
print(LiCnt2)
print(LiCnt3)
```

# append

```python
cnt.append('React')

print(cnt)
```

# insert

```python
cnt.insert(7,'react')

print(cnt)
```

# Append

```python
a = cnt.index('python')

print(a)
```

# remove

```python
cnt.remove('css')

print(cnt)
```

# pop

```python
cnt.pop(7)

print(cnt)
```

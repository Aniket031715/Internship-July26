# Take an file.txt name from user and its contetn
- if a the file already exists do nothing
- if not file should be created and entered the content

```python
filename = input("Please enter the file name : ")

filecon = input("Please enter the file content : ")

myfile = open(filename, "a")
myfile.write(filecon)
myfile.close()
```

```python
myfile2 = open(filename, "r")
pri = myfile2.read()
myfile2.close()

print(pri)
```
